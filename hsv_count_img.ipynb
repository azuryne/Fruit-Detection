```python
# DETECT FROM WEBCAM 
# ADDITIONAL IMAGE PROCESSING 
 
# DETECT FROM WEBCAM
# CROP FROM WEBCAM 

import cv2 as cv
import numpy as np

cap = cv.VideoCapture(0)
  
while True:
    _, frame = cap.read()

    result = frame.copy()
    frame = cv.cvtColor(frame, cv.COLOR_BGR2HSV)

    # lower boundary RED color range values; Hue (0 - 10)
    lower1 = np.array([0, 100, 20])
    upper1 = np.array([10, 255, 255])

    # upper boundary RED color range values; Hue (160 - 180)
    lower2 = np.array([160,100,20])
    upper2 = np.array([179,255,255])
    lower_mask = cv.inRange(frame, lower1, upper1)
    upper_mask = cv.inRange(frame, lower2, upper2)

    full_mask = lower_mask + upper_mask

    result = cv.bitwise_and(result, result, mask=full_mask)

    kernel = np.ones((20, 20), dtype='uint8')

    image_close = cv.morphologyEx(result, cv.MORPH_CLOSE, kernel)

    image1 = cv.cvtColor(image_close, cv.COLOR_HSV2BGR)
    gray = cv.cvtColor(image1, cv.COLOR_BGR2GRAY)

    #gray = cv.cvtColor(, cv.COLOR_BGR2GRAY)
    #_, threshold = cv.threshold(gray, 221, 255, cv.THRESH_BINARY_INV)

    contours, _ = cv.findContours(gray, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

    large = []
    medium = []
    small = []

    for cnt in contours:
        (x, y, w, h) = cv.boundingRect(cnt)

        area = cv.contourArea(cnt)
    
        if area > 1000:
            cv.putText(frame, str(area), (x, y), 2, 1, (255, 0, 0))
            
            if area>40000:
                cv.rectangle(frame, (x, y), (x + w, y + h), (255, 255, 255), 2)
                cv.putText(frame, "LARGE", (x + w, y + h), 2, 1, (255, 255, 255))
                image = frame[y : y + h, x : x + w]
                image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
                large.append(image)


            elif 20000< area <40000:
                cv.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
                cv.putText(frame, "MEDIUM", (x + w, y + h), 2, 1, (0, 255, 0))
                image = frame[y : y + h, x : x + w]
                image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
                medium.append(image)

            else:
                cv.rectangle(frame, (x, y), (x + w, y + h), (123, 25, 79), 2)
                cv.putText(frame, "SMALL", (x + w, y + h), 2, 1, (123, 25, 79))
                image = frame[y : y + h, x : x + w]
                image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
                small.append(image)
        
            cv.putText(frame, f'Large: {str(len(large))}', (50, 50), cv.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 2)
            cv.putText(frame, f'Medium: {str(len(medium))}', (100, 100), cv.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 2)
            cv.putText(frame, f'Small: {str(len(small))}', (150, 150), cv.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 2)



    

    cv.imshow("image",frame)
    cv.imshow("gray", gray)
    # cv.putText(frame, str(len(image_L)), (100, 100), cv.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2)

    
    key = cv.waitKey(10)
    if key == ord("q"):
        break

image_L = np.array(large, dtype='object')
print("Length of large:", len(image_L))

image_M = np.array(medium, dtype='object')
print("Length of medium:",len(image_M))

image_S = np.array(small, dtype='object')
print("Length of small:",len(image_S))

cv.destroyAllWindows()
cv.waitKey(1)
  
```
